```python
#City coordinates

City_Fidel = pd.read_csv("../../Resources/city_coordinates.csv")
City_Fidel = City_Fidel.rename(columns={'lat':'latitude', 'lon':'longitude', 'city':'City'})
City_Fidel
```

```python
#Global Air Pollution

Glob_AP_Nat = pd.read_csv("../../Resources/global air pollution dataset.csv")
Glob_AP_Nat = pd.merge(City_Fidel, Glob_AP_Nat, on='City')
Glob_AP_Nat
```

```python
#Global Land Temp

Glob_Land_Mav = pd.read_csv("../../Resources/Global_Land_Temp_Anom.csv", skiprows=4, delimiter=',')
Glob_Land_Mav['One'] = Glob_Land_Mav['Year'].astype('str')
Glob_Land_Mav['Year'] = Glob_Land_Mav['One'].str[0:4]
Glob_Land_Mav['Month'] = Glob_Land_Mav['One'].str[5:6]
Glob_Land_Mav.drop('One', axis=1, inplace=True)
Glob_Land_Mav = Glob_Land_Mav[['Year', 'Month', 'Anomaly']].set_index('Year')
Glob_Land_Mav
```

```python
#Global Ocean Temp

Glob_Ocean_Mav = pd.read_csv("../../Resources/Global_Ocean_Temp_Anom.csv")
Glob_Ocean_Mav['One'] = Glob_Ocean_Mav['Year'].astype('str')
Glob_Ocean_Mav['Year'] = Glob_Ocean_Mav['One'].str[0:4]
Glob_Ocean_Mav['Month'] = Glob_Ocean_Mav['One'].str[5:6]
Glob_Ocean_Mav.drop('One', axis=1, inplace=True)
Glob_Ocean_Mav = Glob_Ocean_Mav[['Year', 'Month', 'Anomaly']].set_index('Year')
Glob_Ocean_Mav
```

```python
#Sea Level Predictions

Sea_Pre_Jen = pd.read_csv("../../Resources/global_meanslr_predictions.csv")
Sea_Pre_Jen
```

```python
#Sea Level Historical

Sea_Hist_Jen = pd.read_csv("../../Resources/sea-level-rise-historical.csv")
Sea_Hist_Jen = Sea_Hist_Jen.set_index('Year').dropna()
Sea_Hist_Jen
```

```python
#Ocean Warming Data joined with Historical Sea Levels

Ocean_Warm_Mav = pd.read_csv("../../Resources/ocean_warming_data.csv")
Ocean_Warm_Mav = pd.merge(Sea_Hist_Jen, Ocean_Warm_Mav, on='Year')
Ocean_Warm_Mav
```
